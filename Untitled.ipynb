```python
# The PIL (now Pillow) library is more sophisticated and has our drawing functions
#  but a weird image format
# http://www.scipy-lectures.org/advanced/image_processing/

from PIL import Image, ImageDraw

img = Image.open("noggin.bmp")

lineThick = 4

draw = ImageDraw.Draw(img)
draw.line((0, 0) + img.size, fill=128, width=lineThick)
draw.line((0, img.size[1], img.size[0], 0), fill=128, width=lineThick)
del draw

plt.imshow(img)
plt.show()

misc.imsave('nogginCross.bmp', img) # uses the Image module (PIL)
```

```python
# Example marker class - "x marks the measurement point found on the scan"

img = Image.open("noggin.bmp")

class cross:
  def __init__(self, image, position, size, colour, thick):
    self.position = position
    self.size = size
    self.colour = colour
    self.thick = thick
    self.image = image
    
  def drawIt(self):
    self.draw = ImageDraw.Draw(self.image)
    self.draw.line((self.position[0] - int(0.5 + self.size[0]/2), self.position[1], self.position[0] + int(0.5 + self.size[0]/2), self.position[1]), fill=self.colour, width=self.thick)
    self.draw.line((self.position[0], self.position[1] - int(0.5 + self.size[1]/2), self.position[0], self.position[1] + int(0.5 + self.size[1]/2)), fill=self.colour, width=self.thick)
    del self.draw #better in a destruction func?


marker = cross(img, (150,50), (50, 50), 0, 4)

marker.drawIt()

plt.imshow(img)
plt.show()

    
```

```python
# Search for feature over strip of the image
# Crop a strip along our axis of interest
cropped = img.crop((149,0,151,img.size[1]))
plt.imshow(cropped)
plt.title('Cropped strip')
plt.show()

# Turn the strip into a numpy array
import numpy as np
croppedData = np.asarray(cropped, dtype=np.uint8)
#print(croppedData.shape)
#print(croppedData)

# Average the pixels horizontally
averagedData = np.zeros(croppedData.shape[0])

i = 0
for row in croppedData:
  #print('Row: %r Average: %.1f' %(row, np.average(row)))
  averagedData[i] = np.average(row)
  i = i + 1

#print(averagedData.shape)
#print(averagedData)
  
# Plot
plt.plot(averagedData)
plt.title('Averaged colour values - see the vertical black line around 50')
plt.show()

# Convert to RGB
testIm = np.array(cropped.convert("RGB"))
print('PIL format converted to RGB Array')
print(testIm)
```
