```python
# The PIL (now Pillow) library is more sophisticated and has our drawing functions
#  but a weird image format
# http://www.scipy-lectures.org/advanced/image_processing/

from PIL import Image, ImageDraw

img = Image.open("noggin.bmp")

lineThick = 4

draw = ImageDraw.Draw(img)
draw.line((0, 0) + img.size, fill=128, width=lineThick)
draw.line((0, img.size[1], img.size[0], 0), fill=128, width=lineThick)
del draw

plt.imshow(img)
plt.show()

misc.imsave('nogginCross.bmp', img) # uses the Image module (PIL)
```

```python
# Example marker class - "x marks the measurement point found on the scan"

img = Image.open("noggin.bmp")

class cross:
  def __init__(self, image, position, size, colour, thick):
    self.position = position
    self.size = size
    self.colour = colour
    self.thick = thick
    self.image = image
    
  def drawIt(self):
    self.draw = ImageDraw.Draw(self.image)
    self.draw.line((self.position[0] - int(0.5 + self.size[0]/2), self.position[1], self.position[0] + int(0.5 + self.size[0]/2), self.position[1]), fill=self.colour, width=self.thick)
    self.draw.line((self.position[0], self.position[1] - int(0.5 + self.size[1]/2), self.position[0], self.position[1] + int(0.5 + self.size[1]/2)), fill=self.colour, width=self.thick)
    del self.draw #better in a destruction func?


marker = cross(img, (150,50), (50, 50), 0, 4)

marker.drawIt()

plt.imshow(img)
plt.show()

    
```

```python
## Detect RGB colour within range
## probably better to work in HSV encoding but RGB is easier to understand

import numpy as np

myArray = np.zeros(3)

myArray[0]= 140
myArray[1]= 190
myArray[2]= 100

print("Test colour: " + str(myArray))

myRGBUpper = np.zeros(3)
myRGBLower = np.zeros(3)

myRGBUpper[0] = 200
myRGBUpper[1] = 200
myRGBUpper[2] = 200


myRGBLower[0] = 90
myRGBLower[1] = 90
myRGBLower[2] = 90

# Element by element comparison
myOutput = True
for i in range(3):
    if myArray[i] <= myRGBLower[i] or myArray[i] >= myRGBUpper[i]:
        myOutput = False
        break
        
print("Upper limit: " + str(myRGBUpper))
print("Lower limit: " + str(myRGBLower))
print("Within limits? elementwise comparison loop: " + str(myOutput))

# Array comparison
myOutput = all(myArray > myRGBLower) and all(myArray < myRGBUpper) 
print("Within limits? array comparison with '<>' : " + str(myOutput))

myOutput = all(np.greater(myArray,myRGBLower)) and all(np.less(myArray,myRGBUpper))
print("Within limits? array comparison with numpy functions: " + str(myOutput))
```
